### Упражнение 1

Построим линейный и точечный графики для данных, представленных ниже.

Для точечного графика помимо значений величины, график которой нужно построить (y), еще необходимо указать зависимую от нее величину (x). Поскольку нам она не дана в виде исходных данных, ее можно задать как целые число от 0 (или 1) до длины списка представленных данных.

```python
data = [10, 54, 47, 65, 18, 22, 15, 1]
x = list(range(len(data)))

# Линейный график
plt.figure(figsize=(8, 4))
plt.plot(x, data, marker='o', linestyle='-', color='b')
plt.title('Линейный график')
plt.xlabel('Индексы элементов')
plt.ylabel('Значения')
plt.grid(True)
plt.show()

# Точечный график
plt.figure(figsize=(8, 4))
plt.scatter(x, data, color='r', marker='o')
plt.title('Точечный график')
plt.xlabel('Индексы элементов')
plt.ylabel('Значения')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
file_path = "/content/wfp_food_prices_rus.csv"
df = pd.read_csv('/content/wfp_food_prices_rus.csv')
df.head()
```

В датафрейме содержится информация по нескольким регионам. Давайте выделим отдельно данные по Ростовской области и Санкт-Петербургу.

```python
df[df['admin1'].isin(['Rostovskaya Oblast', 'Sankt-peterburg'])]
```

Давайте построим линейные графики цен на зерно в рублях для Ростовской области и Санкт-Петербурга. В качестве аргументов функции plot() используем столбцы с датами и ценами.

```python
df.columns
```

```python
import matplotlib.pyplot as plt
data = pd.read_csv('wfp_food_prices_rus.csv')
rostov_data = data[data['admin1'] == 'Rostovskaya Oblast']
spb_data = data[data['admin1'] == 'Sankt-peterburg']
plt.plot(rostov_data['date'], rostov_data['price'], label='Rostovskaya Oblast')
plt.plot(spb_data['date'], spb_data['price'], label='Sankt-peterburg')
plt.xlabel('Месяц')
plt.ylabel('Цена в рублях')
plt.title('Цены на зерно в Ростовской области и Санкт-Петербурге');
```

### Упражнение 3

Если данные из столбцов не преобразовывать, то окажутся строки и по оси Х, и по оси У. Чтобы это исправить, преобразуйте столбец цен во [float](https://datatofish.com/convert-string-to-float-dataframe/), а столбец с датами [в формат дат](https://datatofish.com/strings-to-datetime-pandas/).

```python
data = {'date': ['2023-01-01', '2023-02-01', '2023-03-01', '2023-04-01'],
        'price': ['10.5', '20.25', '15.75', '30.0']}

df = pd.DataFrame(data)

df['price'] = df['price'].astype(float)
df['date'] = pd.to_datetime(df['date'])

print(df)
```

### Упражнение 4

Добавьте подписи величин по осям Х и У, а также заголовок диаграммы.

```python
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['price'], marker='o', linestyle='-')
plt.title('Динамика цен на пшеницу в России')
plt.xlabel('Дата')
plt.ylabel('Цена')
plt.grid(True)
```

### Упражнение 5
Добавьте легенду к графикам. Также измените тип маркеров для графиков.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('wfp_food_prices_rus.csv')

plt.plot(df['date'], df['price'], marker='o', linestyle='-', label='Цена пшеницы')

plt.legend()

plt.title('Динамика цен на пшеницу в России')
plt.xlabel('Дата')
plt.ylabel('Цена')
plt.grid(True)
```
